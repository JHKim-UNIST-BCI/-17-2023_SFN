```python
import numpy as np
import matplotlib.pyplot as plt


def izhikevich_neuron(a, b, c, d, I, v_init, u_init, time):
    dt = 0.1
    v = v_init
    u = u_init
    v_values = []

    for t in time:
        if v >= 30:
            v = c
            u += d

        dv_dt = 0.04 * v**2 + 5 * v + 140 - u + I
        du_dt = a * (b * v - u)

        v += dv_dt * dt
        u += du_dt * dt
        v_values.append(v)

    return v_values


# Izhikevich neuron parameters
a = 0.02
b = 0.2
c = -65
d = 8

# Initial membrane potential and recovery variable
v_init = -65
u_init = b * v_init

# Simulation time and negative input current
time = np.arange(0, 1000, 0.1)
input_current = 120

# Simulate the neuron and plot the voltage over time
v_values = izhikevich_neuron(a, b, c, d, input_current, v_init, u_init, time)
plt.plot(time, v_values)
plt.xlabel("Time (ms)")
plt.ylabel("Membrane Potential (mV)")
plt.title("Izhikevich Neuron with Negative Input")
plt.show()
```
